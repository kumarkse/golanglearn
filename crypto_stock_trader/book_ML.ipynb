```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,50,100).round(2)
y = np.random.normal(0,20,size=len(x)).round(2) + 2*x +50
# y = 
```

```python
plt.figure(figsize=(14,6))
plt.scatter(x=x,y=y)
plt.grid()
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(x)
model = sm.OLS(y,X).fit()
print(model.summary())
```

```python
Y_hat= model.predict()
```

```python
Y_hat
```

```python
Beta = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
```

```python
Beta
```

```python
Y = np.dot(X,Beta)
Y
```
